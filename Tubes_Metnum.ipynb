```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Perigee dan Apogee pada tahun 2024
perigee_date = "2024-03-10"
perigee_distance = 356894  # km
apogee_date = "2024-10-03"
apogee_distance = 406516  # km

# Semi-major axis (a) dan eksentrisitas (e) orbit Bulan
a = 384400  # km, semi-major axis
e = 0.0549  # eksentrisitas

# Hitung posisi perigee dan apogee
perigee_x = -a * e + perigee_distance
apogee_x = a * (1 - e) + (apogee_distance - a)

# Plot lingkaran unit untuk menunjukkan orbit elips
theta = np.linspace(0, 2 * np.pi, 1000)
orbit_x = a * (np.cos(theta) - e)
orbit_y = a * np.sqrt(1 - e**2) * np.sin(theta)

# Plot orbit dan posisi Perigee dan Apogee
plt.figure(figsize=(10, 8))

# Plot orbit Bulan
plt.plot(orbit_x, orbit_y, label="Orbit Bulan", linestyle='--')

# Plot posisi Perigee dan Apogee
plt.scatter([perigee_x], [0], color='blue', label='Perigee', zorder=5)
plt.scatter([apogee_x], [0], color='red', label='Apogee', zorder=5)

# Annotasi jarak pada Perigee dan Apogee
plt.annotate(f"Perigee\n{perigee_distance} km\n{perigee_date}",
             xy=(perigee_x, 0), xytext=(perigee_x - 50000, 50000),
             arrowprops=dict(arrowstyle="->", color='blue'),
             fontsize=12, color='blue')

plt.annotate(f"Apogee\n{apogee_distance} km\n{apogee_date}",
             xy=(apogee_x, 0), xytext=(apogee_x + 50000, -50000),
             arrowprops=dict(arrowstyle="->", color='red'),
             fontsize=12, color='red')

# Menandai Bumi di pusat
plt.scatter([0], [0], color='green', label='Bumi', s=100, zorder=10)

# Konfigurasi plot
plt.xlabel('Jarak dalam km (sumbu semi-major)')
plt.ylabel('Jarak dalam km (sumbu semi-minor)')
plt.title('Posisi Perigee dan Apogee Bulan pada Tahun 2024')
plt.legend()
plt.grid(True)
plt.axis('equal')

# Menampilkan plot
plt.show()
```

## Perigee using RK-4

```python
# Konstanta dan parameter
G = 6.67430e-11  # Konstanta gravitasi universal, m^3 kg^-1 s^-2
M = 5.972e24     # Massa Bumi, kg
dt = 3600        # Langkah waktu, s (1 jam)
days_per_month = 30  # Rata-rata hari per bulan untuk penyederhanaan
total_months = 12

# Fungsi untuk menghitung percepatan gravitasi
def grav_acc(r):
    r_norm = np.linalg.norm(r)
    return -G * M / r_norm**3 * r

# Inisialisasi posisi dan kecepatan awal Bulan (contoh)
initial_r = np.array([384400e3, 0, 0], dtype=np.float64)  # Posisi awal Bulan, m (jarak rata-rata Bumi-Bulan)
initial_v = np.array([0, 1022, 0], dtype=np.float64)      # Kecepatan awal Bulan, m/s (kecepatan rata-rata orbit)

# Array untuk menyimpan hasil perigee bulanan
perigees = []
dates_of_perigee = []

# Posisi dan kecepatan awal untuk simulasi
r = initial_r.copy()
v = initial_v.copy()

# Loop untuk setiap bulan
for month in range(total_months):
    positions = []
    times = []
    time = 0
    total_time = days_per_month * 86400  # Total waktu simulasi dalam detik (1 bulan)

    while time < total_time:
        positions.append(r.copy())
        times.append(time)
        
        k1_v = grav_acc(r)
        k1_r = v
        k2_v = grav_acc(r + 0.5 * dt * k1_r)
        k2_r = v + 0.5 * dt * k1_v
        k3_v = grav_acc(r + 0.5 * dt * k2_r)
        k3_r = v + 0.5 * dt * k2_v
        k4_v = grav_acc(r + dt * k3_r)
        k4_r = v + dt * k3_v
        
        r += dt / 6 * (k1_r + 2 * k2_r + 2 * k3_r + k4_r)
        v += dt / 6 * (k1_v + 2 * k2_v + 2 * k3_v + k4_v)
        
        time += dt

    positions = np.array(positions)
    distances = np.linalg.norm(positions, axis=1)
    
    # Menemukan perigee untuk bulan ini
    min_distance = np.min(distances)
    perigee_index = np.argmin(distances)
    
    perigees.append(min_distance / 1000)  # Konversi ke km
    
    # Menghitung tanggal spesifik perigee
    days = times[perigee_index] / 86400  # Konversi waktu perigee ke hari
    date = month * days_per_month + days  # Tanggal dalam format hari dari awal tahun
    dates_of_perigee.append(date)

# Visualisasi hasil perigee bulanan
plt.figure(figsize=(10, 5))
months = np.arange(1, total_months + 1)
plt.plot(months, perigees, marker='o', linestyle='-', label='Perigee Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Jarak Perigee (km)')
plt.title('Perigee Bulanan Bulan selama 12 Bulan')
plt.legend()
plt.grid(True)
plt.ylim(356000, 407000)  # Set scale based on realistic Moon-Earth distance
plt.show()

# Cetak hasil perigee bulanan dengan tanggal spesifik
for i, (perigee, date_of_perigee) in enumerate(zip(perigees, dates_of_perigee), start=1):
    month_day = int(date_of_perigee)
    hour_of_day = (date_of_perigee % 1) * 24
    print(f"Bulan {i}: Perigee terjadi pada hari ke-{month_day} dengan jarak {perigee:.2f} km, sekitar pukul {hour_of_day:.2f}")
```
